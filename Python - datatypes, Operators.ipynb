### Python program using strings, numbers, datetime objects, we will display datatypes

```python
# string 
name = input('Enter name: ')
print('Name of the user is {0}. The datatype is{1} '.format(name, type(name)))

# integer
age = input('Enter age: ')
print('Age datatype before conversion is ',type(age))
age = int(age)
print('Age of the user is {0}. The datatype after conversion is {1} '.format(str(age), type(age)))

# datetime
dob = input('Enter Date of Birth (format: jan 01 2000 12:00 AM): ')
print('DOB datatype before conversion is ',type(dob))

#converting date string into datetime format
from datetime import datetime
dob = datetime.strptime(dob,'%b %d %Y %I:%M %p')
print('DOB of the user is {0}. The datatype after conversion is {1} '.format(str(dob), type(dob)))
```

In python all inputs are considered as string by default. In the above program if we check the datatype of name, age and date we see that all are in string format. <br>
We can convert Age into integer format using int() bulint-in function. Which we have done in the program. <br>
To convert Date into datetime format we need to install datetime library and use datetime() function. <br>
Another interesting builtin function is strptime() which lets us choose the style of date format.

### Python program to generate list of random integers, display their sum, maximum and minimum.

```python
#Generating random integers

import random
from random import seed
from random import randint

#seeding the random number generator to make list constant 
seed(1)   

#creating empty list
random_list = []

#generating 5 random integers from -10 to 10
for i in range(5):
    n = random.randint(-10,10)
    #adding random integers into the list random_list
    random_list.append(n)


print(random_list)
print('Sum of random numbers is:', sum(random_list))
print('Maximum value from random numbers is:', max(random_list))
print('Minimum value from random numbers is:', min(random_list))
```

### Using sample() function to select random integers, by using range() function you can select random numbers from between a range of numbers.

```python
#Generating random numbers

import random
#from random import seed

#seeding the random number generator to make list constant 
seed(1)   

#creating list of 5 random numbers
random_list = random.sample(range(0,10),5)

print(random_list)
print('Sum of random numbers is:', sum(random_list))
print('Maximum value from random numbers is:', max(random_list))
print('Minimum value from random numbers is:', min(random_list))
```

 ### Use of 'is' operator in List, Tuple, Set, Dictionary

```python
# Is operator in List
list1 = [1,2]
list2 = [1,2]
print(list2 is list1)  #returns false because different memory is allocated to each list value

# Is operator in tuple
tuple1 = (1)
tuple2 = (1)
tuple3 = (1,2,3)
tuple4 = (1,2,3)
print(tuple1 is tuple2) #returns true for single value
print(tuple3 is tuple4) #returs false for more than 1 value as memory location varies

# Is operator in set
set1 = {1,2,3,4}
set2 = {1,2,3,4}
print(set1 is set2)

# Is operator in dictionary
dict1 = {1:'a'}
dict2 = {1:'a'}
print(dict1 is dict2)
```

 ### Use of 'is' operator in List, Tuple, Set, Dictionary

```python
# In operator in list
x = [1,2,3,4,5,6]
y = 2
z = [1,2]
w = ["Hello"]
print(y in x) #prints true since both are in same memory address
print(z in x) #prints false because 2 separate list have different memory address
print("hello" in w) #prints false as python is case sensitive

# In operator in tuple
x = (9,8,7,6)
y = 9
z = (9,8,7,6)
print(y in x) #prints true since both are in same memory address
print(z in x) #prints false because 2 separate tuple have different memory address

# In operator in set
x = {(1,2), (3,4)}
y = (3,4)
z = 2
print(y in x) #returns true because both variables are in same memory address
print(z in x) #returns false z = 2 is a separate variable and not in the dictionary
```
